```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df = df.iloc[:, 2:]
df.sample(5)
```

```python
# --- Scatter Plot: Original Data ---
plt.figure(figsize=(6, 5))
plt.scatter(df[df.columns[0]],df[df.columns[1]], alpha=0.7)
plt.title("Original Data")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
centered = df - df.mean()
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(centered[centered.columns[0]],centered[centered.columns[1]], alpha=0.7)
plt.title("Centered Data")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("Purchased", axis= 1), df["Purchased"], test_size = 0.2, stratify=df["Purchased"])
X_train, X_test, y_train, y_test
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()

scalar.fit(X_train)

X_train_scaled = scalar.transform(X_train)
X_test_scaled = scalar.transform(X_test)

# if not done the above thing => Model was trained in one “scale-space” but tested in another → inconsistent, leads to poor performance.
# predict in the same space as the training data as this wil help in identifying the patterns etc quickly...

X_train_scaled.shape
```

```python
scalar.mean_
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
X_train_scaled
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)

# after scaling
ax2.set_title('Age Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('Salary Distribution Standard Scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

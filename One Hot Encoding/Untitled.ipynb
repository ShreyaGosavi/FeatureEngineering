## One Hot Encoding using sklearn

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 0:4], df.iloc[:, -1], test_size = 0.2, random_state = 0)
X_train.sample(5)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_train_new.shape
```

```python
X_train_new = np.hstack((X_train[['brand','km_driven']].values,X_train_new))
X_train_new
```

```python
X_train_new = pd.DataFrame(X_train_new, columns = ['brand', 'km_driven', 'fuel_Diesel', 'fuel_LPG', 'fuel_Petrol','owner_Fourth & Above Owner', 'owner_Second', 'owner_Test Drive Car',  'owner_Third Owner' ])
X_train_new
```

## OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].nunique()
```

```python
threshold = 100
```

```python
repl = counts[counts <= threshold].
#It identifies the categories in a column that occur less than or equal to a specified threshold,
often used to group rare categories into a single label like "Other" before encoding.
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon')).sample(5)
```
